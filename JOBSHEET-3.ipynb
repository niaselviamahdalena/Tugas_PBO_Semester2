# Praktikum 01 Enkapsulasi – BankAccount

```python
class BankAccount:

    # Konstruktor
    def __init__(self, owner, balance):
        self.__owner = owner
        self.__balance = balance

    # Method deposit
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"{amount} telah ditambahkan ke akun {self.__owner}.")
        else:
            print("Jumlah deposit harus lebih dari 0.")

    # Method withdraw
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"{amount} telah ditarik dari akun {self.__owner}.")
        else:
            print("Saldo tidak mencukupi.")

    # Method get_balance
    def get_balance(self):
        return self.__balance


# Contoh penggunaan
if __name__ == "__main__":

    alice_account = BankAccount("Alice", 1000)

    alice_account.deposit(500)
    alice_account.deposit(-100)

    alice_account.withdraw(300)
    alice_account.withdraw(2000)

    current_balance = alice_account.get_balance()
    print("Saldo terakhir:", current_balance)
```

# Praktikum 02 Enkapsulasi – Employee

```python
class Employee:

    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    def update_salary(self, increase):
        if increase > 0:
            self.__salary += increase

    def set_salary(self, new_salary):
        if new_salary >= 0:
            self.__salary = new_salary

    def get_salary(self):
        return self.__salary

    def get_employee_info(self):
        print("Nama:", self.__name)
        print("Gaji:", self.__salary)


if __name__ == "__main__":
    emp = Employee("Budi", 5000000)

    emp.update_salary(1000000)
    emp.get_employee_info()
```

# Praktikum 04 FileLogger (Constructor & Destructor)

```python
class FileLogger:

    def __init__(self, filename):
        self.filename = filename
        self.file = open(filename, "a")
        print(f"File '{filename}' berhasil dibuka untuk logging.")

    def write_log(self, message):
        self.file.write(message + "\n")
        self.file.flush()
        print(f"Pesan log: '{message}' telah ditulis.")

    def __del__(self):
        self.file.close()
        print("File log ditutup.")


if __name__ == "__main__":
    logger = FileLogger("log.txt")
    logger.write_log("Program dimulai")
    del logger
```

# Praktikum 05 Peran self (Calculator)

```python
class Calculator:

    def __init__(self, initial_value=0):
        self.value = initial_value

    def add(self, number):
        self.value += number
        print("Setelah penambahan:", self.value)

    def subtract(self, number):
        self.value -= number
        print("Setelah pengurangan:", self.value)

    def reset(self):
        self.value = 0
        print("Nilai telah direset ke 0.")

    def show_value(self):
        print("Nilai saat ini:", self.value)


def main():
    calc1 = Calculator(10)
    calc1.add(5)
    calc1.subtract(3)
    calc1.show_value()

    calc2 = Calculator()
    calc2.add(20)
    calc2.show_value()


if __name__ == "__main__":
    main()
```

# Praktikum 06 Menggunakan getter, setter, dan dekorator @property untuk mengelola akses atribut

```python
class Person:

    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        if value != "":
            self.__name = value

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if value >= 0:
            self.__age = value


if __name__ == "__main__":
    person = Person("Nia", 18)

    print(person.name, person.age)

    person.age = 20
    print("Umur baru:", person.age)
```

```python
class Student:
    # Konstruktor
    def __init__(self, name, score):
        self.__name = name      # atribut privat
        self.__score = 0
        self.__grade = "E"
        self.score = score      # pakai setter agar tervalidasi

    # ===== Getter & Setter NAME =====
    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        self.__name = value

    # ===== Getter & Setter SCORE =====
    @property
    def score(self):
        return self.__score

    @score.setter
    def score(self, value):
        if 0 <= value <= 100:
            self.__score = value
            self.__update_grade()
        else:
            print("Nilai harus antara 0 - 100!")

    # ===== Method internal update grade =====
    def __update_grade(self):
        if self.__score >= 90:
            self.__grade = "A"
        elif self.__score >= 80:
            self.__grade = "B"
        elif self.__score >= 70:
            self.__grade = "C"
        elif self.__score >= 60:
            self.__grade = "D"
        else:
            self.__grade = "E"

    # ===== Method tampilkan data =====
    def show_info(self):
        print("Nama Mahasiswa :", self.__name)
        print("Nilai          :", self.__score)
        print("Grade          :", self.__grade)

    # ===== Destruktor =====
    def __del__(self):
        print(f"Data mahasiswa {self.__name} telah dihapus dari sistem.")


# ===============================
# Program Utama
# ===============================
nama = input("Masukkan nama mahasiswa : ")
nilai = int(input("Masukkan nilai mahasiswa : "))

mhs = Student(nama, nilai)

print("\n=== DATA MAHASISWA ===")
mhs.show_info()

del mhs
```

```python
class Student:
    # Konstruktor
    def __init__(self, name, score):
        self.__name = name
        self.__score = 0
        self.__grade = "E"
        self.score = score  # lewat setter

    # Getter & Setter NAME
    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        self.__name = value

    # Getter & Setter SCORE
    @property
    def score(self):
        return self.__score

    @score.setter
    def score(self, value):
        if 0 <= value <= 100:
            self.__score = value
            self.__update_grade()
        else:
            print("Nilai harus 0 - 100")

    # Update Grade Otomatis
    def __update_grade(self):
        if self.__score >= 90:
            self.__grade = "A"
        elif self.__score >= 80:
            self.__grade = "B"
        elif self.__score >= 70:
            self.__grade = "C"
        elif self.__score >= 60:
            self.__grade = "D"
        else:
            self.__grade = "E"

    # Tampilkan Info
    def show_info(self):
        print(f"Nama Mahasiswa: {self.__name} Nilai: {self.__score}")
        print(f"Grade: {self.__grade}\n")

    # Destruktor
    def __del__(self):
        print(f"Data mahasiswa {self.__name} telah dihapus dari sistem.")

# PROGRAM UTAMA

mhs = Student("Nia", 85)

# Output pertama
mhs.show_info()

# Ubah nilai
print("Nilai diubah...")
mhs.score = 98

# Output kedua
mhs.show_info()

# Hapus object
del mhs
```
