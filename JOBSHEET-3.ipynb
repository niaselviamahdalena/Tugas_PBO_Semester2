## D. Langkah Praktikum

### Praktikum 01: Enkapsulasi pada kelas BankAccount
- Atribut privat: `__owner` dan `__balance`.
- Metode publik: `__init__`, `deposit(amount)`, `withdraw(amount)`, dan `get_balance()`.
- `deposit()` hanya menerima nilai positif dan menambah saldo jika valid.
- `withdraw()` melakukan validasi agar penarikan tidak melebihi saldo.
- Atribut privat di Python menggunakan name mangling (`__balance` menjadi `_BankAccount__balance`), yang menandakan bahwa atribut ini tidak seharusnya diakses langsung.

### Praktikum 02: Enkapsulasi pada kelas Employee
- Atribut privat: `__name` dan `__salary`.
- Metode publik: `__init__`, `update_salary(increase)`, `set_salary(new_salary)`, `get_salary()`, dan `get_employee_info()`.
- Validasi gaji memastikan nilai tidak negatif saat diubah.
- Data gaji hanya diakses dan dimodifikasi melalui metode yang disediakan.

### Praktikum 03: Konstruktor dan Destruktor Sederhana
- Atribut: `name` sebagai data publik.
- Metode: `__init__(name)` untuk inisialisasi, `__del__()` untuk destruktor.
- `__init__` dipanggil saat objek dibuat.
- `__del__` dipanggil ketika objek dihapus atau garbage collector membersihkannya.

### Praktikum 04: Konstruktor dan Destruktor program FileLogger
- Atribut: `filename` dan `file` (objek file yang dibuka dalam mode append).
- Metode: `__init__(filename)` membuka file log, `write_log(message)` menulis log, dan `__del__()` menutup file saat objek dihapus.
- Contoh ini menunjukkan pengelolaan resource: file harus ditutup sehingga tidak bocor.

### Praktikum 05: Properti `self` pada kelas
- Atribut: `value` pada kelas `Calculator`.
- Metode: `__init__(initial_value=0)`, `add(number)`, `subtract(number)`, `reset()`, dan `show_value()`.
- `self` merujuk pada instance saat ini sehingga setiap objek mempunyai nilai dan perilaku sendiri.
- Perubahan pada satu objek tidak memengaruhi objek lain.

### Praktikum 06: Getter, Setter, dan Dekorator `@property`
- Atribut privat: `__name` dan `__age` pada kelas `Person`.
- Properti: `@property` untuk getter, `@<property>.setter` untuk setter.
- Validasi: nama tidak boleh kosong, usia tidak boleh negatif.
- Akses properti tetap seperti atribut biasa, tetapi logika validasi dijalankan di balik layar.

## G. Kesimpulan
Praktikum ini mengajarkan konsep dasar OOP di Python secara praktis. Mahasiswa belajar:
- Enkapsulasi untuk melindungi data dan menambahkan validasi.
- Konstruktor `__init__` untuk inisialisasi dan destruktor `__del__` untuk membersihkan resource.
- Peran `self` untuk mengelola atribut instance.
- Getter, setter, dan dekorator `@property` untuk akses atribut yang lebih aman.

Dengan demikian, mahasiswa dapat membangun program yang lebih terstruktur, terenkapsulasi, dan lebih mudah dipelihara.

```python
class Student:
    def __init__(self, name: str, score: int):
        self._name = None
        self._score = None
        self._grade = None
        self.name = name
        self.score = score

    @property
    def name(self) -> str:
        return self._name

    @name.setter
    def name(self, value: str):
        if not isinstance(value, str) or not value.strip():
            raise ValueError("Nama harus berupa teks tidak kosong.")
        self._name = value.strip()

    @property
    def score(self) -> int:
        return self._score

    @score.setter
    def score(self, value: int):
        if not isinstance(value, int):
            raise ValueError("Score harus berupa bilangan bulat.")
        if value < 0 or value > 100:
            raise ValueError("Score harus berada di antara 0 dan 100.")
        self._score = value
        self._grade = self._calculate_grade()

    @property
    def grade(self) -> str:
        return self._grade

    def _calculate_grade(self) -> str:
        if self._score >= 90:
            return "A"
        if self._score >= 80:
            return "B"
        if self._score >= 70:
            return "C"
        if self._score >= 60:
            return "D"
        return "E"

    def show_info(self) -> None:
        print(f"Nama Mahasiswa: {self.name}")
        print(f"Nilai: {self.score}")
        print(f"Grade: {self.grade}")

    def __del__(self):
        print(f"Data mahasiswa {self.name} telah dihapus dari sistem.")


# Contoh penggunaan
if __name__ == "__main__":
    student = Student("Andi", 85)
    student.show_info()
    print()

    student.score = 92
    student.show_info()
    print()

    del student
```
